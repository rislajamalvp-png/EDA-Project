## **Hypothesis Testing**

**Do Tier 1 and Tier 3 city people have significantly different savings rates — or is it just by chance?**

**1. T-Test — Savings Rate by City Tier**


```python
from scipy import stats
tier1=df[df['City_Tier']=='Tier_1']['Savings_Rate']
tier2=df[df['City_Tier']=='Tier_2']['Savings_Rate']

t_stat, p_value=stats.ttest_ind(tier1,tier2)
print(f"T-statistic:{t_stat:.4f}")
print(f"P-value:{p_value:.4f}")

if p_value<0.05:
    print('Statistically Significant - City Tier does affect Savings rate')
else:
    print('Due to some random chance - City Tier does not affect Savings rate ')
```



### 1. T-Test — Savings Rate by City Tier

**Problem Statement:**
Do people living in different city tiers have significantly different savings rates?

**Why this matters:**
Our boxplot showed Tier 3 saves more than Tier 1 — but is this difference 
real or just random chance? T-test confirms this statistically.

**Hypothesis:**
- H0 (Null): There is NO significant difference in savings rate 
  between Tier 1 and Tier 3 city people
- H1 (Alternate): There IS a significant difference in savings rate 
  between Tier 1 and Tier 3 city people

**Significance Level:** α = 0.05

**Result:**
- T-statistic: -76.8626
- P-value: 0.0000

**Conclusion:**
Since p-value (0.0000) < 0.05, we REJECT the null hypothesis.
✅ City Tier significantly affects Savings Rate — Tier 3 people 
save significantly more than Tier 1 people.

**Anova**

**Does Income Segment significantly affect Savings Rate?**

```python
from scipy import stats
low=df[df['Income_Segment']=='low']['Savings_Rate']
middle=df[df['Income_Segment']=='middle']['Savings_Rate']
high=df[df['Income_Segment']=='high']['Savings_Rate']
very_high=df[df['Income_Segment']=='very_high']['Savings_Rate']

f_stat,p_value=stats.f_oneway(low,middle,high,very_high)

print(f"F-statistic:{f_stat:.4f}")
print(f"P-value:{p_value:.4f}")

if p_value<0.05:
    print('statisticaly significant-Income Segment DOES affect Savings Rate')
else:
    print('Due to random chance-Income segement does not affect savings')
```

### 2. ANOVA — Income Segment vs Savings Rate

**Problem Statement:**
Does a person's income level significantly affect their savings rate?

**Hypothesis:**
- H0 (Null): There is NO significant difference in savings rate 
  across income segments
- H1 (Alternate): There IS a significant difference in savings rate 
  across income segments

**Significance Level:** α = 0.05

**Result:**
- F-statistic: 1.8738
- P-value: 0.1316

**Conclusion:**
- Since p-value (0.1316) > 0.05, we FAIL TO REJECT the null hypothesis.
- Income segment does NOT significantly affect savings rate — 
- people save at similar rates regardless of how much they earn.

```python
from scipy.stats import chi2_contingency

contingency_table=pd.crosstab(df['City_Tier'],df['Occupation'])
print(contingency_table)

chi2, p_value, dof, expected= chi2_contingency(contingency_table)

print(f" \nChi_square Statistic:{chi2:.4f}")
print(f"P-value:{p_value:.4f}")
print(f"Degrees of freedom:{dof}")

if p_value<0.05:
    print("statisticaly significant- City Tier and Occupation are related")
else:
    print('Due to some random chance - City Tier and Occupation are not related')
```

### 3. Chi-Square — City Tier vs Occupation

**Problem Statement:**
Is there a relationship between which city tier a person lives in 
and their occupation?

**Hypothesis:**
- H0 (Null): City Tier and Occupation are NOT related
- H1 (Alternate): City Tier and Occupation ARE related

**Significance Level:** α = 0.05

**Result:**
- Chi-square Statistic: 1.6340
- P-value: 0.9501
- Degrees of Freedom: 6

**Conclusion:**
- Since p-value (0.9501) > 0.05, we **FAIL TO REJECT** the null hypothesis.
- No significant relationship — City Tier and Occupation are 
- independent of each other.

*Each occupation category contains nearly the same number of individuals, allowing fair comparison without group imbalance affecting the results.*

#### **5: Higher Income Leads to Higher Spending** 
*Income and expenses move together strongly — when earnings increase, spending also increases at a similar pace, showing lifestyle expansion rather than increased saving.*

*There is almost no correlation between age and income, expenses, or savings. Financial patterns appear lifestyle-based rather than age-based.*

#### **7: Dependents Reduce Ability to Save**
*Individuals with more dependents tend to save less, as household responsibilities reduce the remaining disposable income.*

#### **8: Tier 1 Residents Save the Least**
*People living in Tier 1 cities have the lowest median savings rate (around 19%), likely due to higher cost of living pressures.*

#### **9: Tier 3 Residents Save the Most**
*People in Tier 3 cities achieve the highest savings rate (around 33%), benefiting from lower living expenses and better surplus income.*

#### **10: Location Significantly Impacts Savings**
*Statistical testing confirms that savings differences between city tiers are significant, meaning location genuinely affects financial behaviour.*

#### **11: Income Level Does Not Change Saving Habit**
*Across different income groups, savings rates remain similar, suggesting people adjust their lifestyle as income grows rather than saving a larger percentage.*

#### **12: Occupation and City Tier Are Unrelated**
